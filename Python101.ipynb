```python
#final exam 150 question, pass >=120
def grade(score):
    if score >= 120:
      return "Excellent"
    elif score >= 100:
      return "Good"
    elif score >= 80:
      return "Okay"
    else:
      return "Need to read more!"

grade(95)
```

```python
#use and, or in condition
#course == data science, score >= 80 passed
#course == english, score >= 70 passed
def grade1(course, score):
  if course == "english" and score >= 70:
    return "passed"
  elif course == "data science" and score >= 80:
    return "passed"
  else:
    return "failed"
```

```python
grade1("data science", 75)
```

```python
#for loop
#if score >= 80, passed
def grading_all(scores):

    new_scores = []

    for score in scores:
      new_scores.append(score+2)
    return new_scores
```

```python
def chatbot():
  fruits = []
  while True:
    fruit = input("What fruit do you want to order? ")
    if fruit not in ["exit", "quit"]:
      fruits.append(fruit)
    else:
      break
  return(fruits)
```
